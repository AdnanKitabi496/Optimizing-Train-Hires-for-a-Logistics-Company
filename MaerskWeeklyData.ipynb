```python
import pandas as pd
import numpy as np
```

```python
weeks = 46
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
strike_weeks = [5, 39, 40, 44, 45]
cols = ['Terminal', 'Type', 'Head Code', 'Depart Time', 'Core Wagons',
       'Laden Containers - 20ft', 'Laden Containers - 40ft', 'Laden Containers - HC', 'Laden Containers - 45ft',
       'Empty Containers - 20ft', 'Empty Containers - 40ft', 'Empty Containers - HC', 'Empty Containers - 45ft',
       'Addnl Wagons', 'Buy Back - 20ft', 'Buy Back - 40ft', 'Available TEU', 'Used TEU',
       'Utilisation', 'Total Conts']
```

```python
data = []
for week in range(weeks):
    if week + 1 not in strike_weeks:
        filename_felixstowe = "Maersk_Data/WEEK" + str(week + 1).zfill(2) + "2022.xlsm"
        filename_londongateway = "Maersk_Data/WEEK" + str(week + 1).zfill(2) + "2022LGW.xlsx"
        for day in days:
            # Loop through Felixstowe Days in Week
            df = pd.read_excel(filename_felixstowe, sheet_name=day, skiprows=9, usecols="A:T", header=None, names=cols)
            import_row = df.index[df["Terminal"] == 'TOTAL IMPORTS'].to_list()[0]
            df_import = df[:import_row-1]
            df_import['Day'] = day
            df_import['Week'] = week + 1
            df_import['Direction'] = 'Import'
            df_import['Port'] = 'Felixstowe'
            data.append(df_import)
            export_row = df.index[df["Terminal"] == 'TOTAL EXPORTS'].to_list()[0]
            df_export = df[import_row+10:export_row-1]
            df_export['Day'] = day
            df_export['Week'] = week + 1
            df_export['Direction'] = 'Export'
            df_export['Port'] = 'Felixstowe'
            data.append(df_export)
            # Loop through LGW Days in Week
            df = pd.read_excel(filename_londongateway, sheet_name=day, skiprows=9, usecols="A:T", header=None, names=cols)
            import_row = df.index[df["Terminal"] == 'TOTAL IMPORTS'].to_list()[0]
            df_import = df[:import_row-1]
            df_import['Day'] = day
            df_import['Week'] = week + 1
            df_import['Direction'] = 'Import'
            df_import['Port'] = 'London Gateway'
            data.append(df_import)
            export_row = df.index[df["Terminal"] == 'TOTAL EXPORTS'].to_list()[0]
            df_export = df[import_row+10:export_row-1]
            df_export['Day'] = day
            df_export['Week'] = week + 1
            df_export['Direction'] = 'Export'
            df_export['Port'] = 'London Gateway'
            data.append(df_export)
df = pd.concat(data)
#df.to_csv('clean.csv', index= False)
```

```python
df2 = df
df2['Used_TEU_Before_BuyBack'] = df2['Used TEU'] - df2['Buy Back - 20ft'] - df2['Buy Back - 40ft'] * 2
df2[df2['Buy Back - 40ft'] > 10]
```

```python
aggregate = df2.groupby(['Head Code', 'Port', 'Terminal']).agg({'Available TEU': 'sum', 'Used_TEU_Before_BuyBack': 'sum', 'Used TEU': 'sum'})
aggregate['Available TEU'] = aggregate['Available TEU'].astype('float')
aggregate['Used_TEU_Before_BuyBack'] = aggregate['Used_TEU_Before_BuyBack'].astype('float')
aggregate['Utilisation_Before_BuyBack'] = aggregate['Used_TEU_Before_BuyBack'].divide(aggregate['Available TEU'], fill_value=None)
aggregate

#Cleaning issues - $140 Headcode on Felixstowe-Doncaster
#L at the end of Terminal names
```
