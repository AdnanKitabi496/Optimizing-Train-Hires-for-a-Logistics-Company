```python
import pandas as pd;
import geopandas as gpd;
import matplotlib.pyplot as plt;

fp = "London_Boroughs.gpkg"
map_df = gpd.read_file(fp)
# check data type so we can see that this is not a normal dataframe, but a GEOdataframe
map_df.head()
```

```python
map_df
```

```python
df = pd.read_csv("tenure_data.csv", header=0)
df.head()
```

```python
owners_df = df[(df['Tenure'] == "Own.Outright") & (df['Year'] == 2018) & (df['Code'].str[:3] == "E09")]
```

```python
owner_slice = owners_df[['Code', 'Name', 'Percent_of_households_in_borough']]
data_for_map = owner_slice.rename(index=str, columns={"Code": "code", "Name": "name", "Percent_of_households_in_borough": "owner_pct"})
data_for_map.head()
```

```python
merged = map_df.set_index("name").join(data_for_map.set_index("name"))
merged.head()
```

```python
merged_slice = merged[["geometry", "owner_pct"]]
merged_slice
```

```python
variable = "owner_pct"
vmin, vmax = 0, 40
fig, ax = plt.subplots(1, figsize=(10, 6))
merged.plot(column=variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')
```
