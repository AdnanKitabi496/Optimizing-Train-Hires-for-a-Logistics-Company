```python
#import libraries
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
%matplotlib notebook
pd.set_option('display.precision', 2)
```

```python
#Load Processed North East Distribution Centre Data
df = pd.read_csv(r'DC_data_processed.csv')

#Weight distribution centres by stage of development
existing_weight = 1
pipeline_weight = 0
land_weight = 0

df['wmsf'] = np.where(df['Category'] == 'Occupied', df['msf'] * existing_weight, np.where(df['Category'] == 'Pipeline', df['msf'] * pipeline_weight, df['msf'] * land_weight))

df
```

```python
#Set up terminal options
data = [['FL Leeds', 53.7673, -1.5062, (5,5)], ['FL Doncaster', 53.508157, -1.125958, (5,5)], ['iPort', 53.48247871041433,-1.1087608337402344, (5,-5)], ['Tinsley', 53.4011002,-1.3850275, (5,5)], ['Gascoigne Wood', 53.7780692,-1.2133953,  (5,5)], ['Wakefield', 53.71375,-1.4077798, (5,5)],  ['Teesport', 54.602699,-1.1465941,  (5,5)], ['East Midlands Gateway', 52.8380088,-1.3067932, (5,5)], ['Goole', 53.700448, -0.882403, (5,5)], ['Hatfield', 53.592417, -1.020067, (5,5)]]
fl = [['FL Leeds', 53.7673, -1.5062, (5,5)], ['FL Doncaster', 53.508157, -1.125958, (5,5)]]  
# Create the pandas DataFrame
all_locations = pd.DataFrame(data, columns=['Terminal', 'Lat', 'Long', 'map_offset'])  
all_locations
```

```python
df1 = df
for terminal in all_locations['Terminal']:
    df1[terminal + '_band'] = (df1[terminal + '_distance'] - df1[terminal + '_distance'].mod(10))/10
df1
```

```python
max_band = 10
df2 = pd.DataFrame(index=list(range(max_band+1)))


for terminal in all_locations['Terminal']:
    data=[]
    bands = df1.groupby([terminal + '_band'])['wmsf'].sum()
    for x in range(max_band+1):
        data.append(bands.loc[x])
    df2[terminal] = data

df2

#for item in msf_array:
    
```

```python
df2.to_csv('dc_analysis/DC_Analysis_Output_all.csv')
```

```python
max_band = 3
df3 = pd.DataFrame(index=all_locations['Terminal'].tolist())
catchment_data = []
doncaster_data=[]

for terminal in all_locations['Terminal']:
    catchment = df1[(df1[terminal + '_band'] <= max_band)]['wmsf'].sum() 
    doncaster_unreachable = df1[(df1[terminal + '_band'] <= max_band) & ~(df1['FL Doncaster_band'] <= max_band)]['wmsf'].sum()
    catchment_data.append(catchment)
    doncaster_data.append(doncaster_unreachable)

df3['Catchment_2'] = catchment_data
df3['Doncaster_Unreachable_2'] = doncaster_data
df3['Shared Catchment'] = df3['Catchment_2'] - df3['Doncaster_Unreachable_2']
df3

    
```
